```python
# Conteggio valori mancanti per colonna
missing = df.isnull().sum()

# Mostra solo le colonne che hanno almeno un NaN
missing_cols = missing[missing > 0]

if len(missing_cols) == 0:
    print("✅ Nessun valore mancante nel dataset!")
else:
    print("⚠️ Colonne con valori mancanti:\n")
    # Calcoliamo anche la percentuale
    missing_pct = (missing_cols / len(df) * 100).round(2)
    missing_report = pd.DataFrame({
        'Mancanti': missing_cols,
        'Percentuale (%)': missing_pct
    }).sort_values('Mancanti', ascending=False)
    print(missing_report)
```

```python
# Salviamo i nomi originali per riferimento
print("Colonne originali:")
print(df.columns.tolist())
print()

# Rinominiamo le colonne con nomi più corti e senza spazi
# Mappatura basata sui nomi REALI del nostro CSV
rename_map = {
    'Country name': 'country',
    'Regional indicator': 'region',
    'Life Ladder': 'happiness_score',
    'Log GDP per capita': 'gdp_per_capita',
    'Social support': 'social_support',
    'Healthy life expectancy at birth': 'life_expectancy',
    'Freedom to make life choices': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'corruption',
    'Positive affect': 'positive_affect',
    'Negative affect': 'negative_affect',
}

# Rinomina solo le colonne che esistono nel dataset
actual_renames = {k: v for k, v in rename_map.items() if k in df.columns}
df = df.rename(columns=actual_renames)

print("Colonne rinominate:")
print(df.columns.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Documentazione della dimensione iniziale
initial_shape = df.shape[0]
print(f"Dimensione iniziale del dataset: {initial_shape} record")

# 2. Rimozione degli studenti con target 'Enrolled'
# Filtriamo mantenendo solo chi ha concluso il percorso
df_filtered = df[df['Target'] != 'Enrolled'].copy()

# 3. Codifica numerica del Target
# Dropout = 0, Graduate = 1
df_filtered['Target'] = df_filtered['Target'].map({'Dropout': 0, 'Graduate': 1})

# 4. Calcolo delle statistiche e distribuzione
final_shape = df_filtered.shape[0]
removed_count = initial_shape - final_shape
distribuzione = df_filtered['Target'].value_counts(normalize=True) * 100

print(f"Studenti rimossi ('Enrolled'): {removed_count}")
print(f"Dimensione finale del dataset: {final_shape} record")
print("\nNuova distribuzione del Target:")
print(f"- Graduate (1): {distribuzione[1]:.2f}%")
print(f"- Dropout (0): {distribuzione[0]:.2f}%")

# 5. Visualizzazione (Da salvare per la presentazione)
plt.figure(figsize=(10, 6))

# Subplot 1: Prima della rimozione (Semplificato)
plt.subplot(1, 2, 1)
sns.countplot(x='Target', data=df, palette='viridis')
plt.title('Distribuzione Originale')

# Subplot 2: Dopo la rimozione e codifica
plt.subplot(1, 2, 2)
sns.barplot(x=distribuzione.index, y=distribuzione.values, palette='magma')
plt.xticks([0, 1], ['Dropout (0)', 'Graduate (1)'])
plt.ylabel('Percentuale %')
plt.title('Target Finale (Binary Classification)')

plt.tight_layout()
plt.show()
```

```python
# 1. Pulizia preventiva delle stringhe (rimuove spazi bianchi prima/dopo)
df['Target'] = df['Target'].str.strip()

# 2. Documentazione dimensione iniziale
initial_shape = df.shape[0]

# 3. Rimozione Enrolled
df_filtered = df[df['Target'] != 'Enrolled'].copy()

# 4. Codifica numerica
# Usiamo un dizionario esplicito
mapping = {'Dropout': 0, 'Graduate': 1}
df_filtered['Target'] = df_filtered['Target'].map(mapping)

# 5. Verifica e Stampa
final_shape = df_filtered.shape[0]
distribuzione = df_filtered['Target'].value_counts(normalize=True) * 100

print(f"Record iniziali: {initial_shape}")
print(f"Record rimossi: {initial_shape - final_shape}")

# Usiamo .get() o controlliamo l'esistenza per evitare nuovi KeyError
if not distribuzione.empty:
    grad_perc = distribuzione.get(1, 0)
    drop_perc = distribuzione.get(0, 0)
    print(f"\nNuova distribuzione:")
    print(f"- Graduate (1): {grad_perc:.2f}%")
    print(f"- Dropout (0): {drop_perc:.2f}%")
else:
    print("Errore: Il dataset filtrato è vuoto o il mapping non è riuscito.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepariamo i dati per il confronto
data_before = df['Target'].value_counts()
data_after = df_filtered['Target'].map({1: 'Graduate', 0: 'Dropout'}).value_counts()

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Grafico PRIMA
sns.barplot(x=data_before.index, y=data_before.values, ax=ax[0], palette='viridis')
ax[0].set_title(f'Prima: Dataset Completo\n({initial_shape} record)')
ax[0].set_ylabel('Numero di Studenti')

# Grafico DOPO
sns.barplot(x=data_after.index, y=data_after.values, ax=ax[1], palette='magma')
ax[1].set_title(f'Dopo: Solo Outcome Definitivi\n({final_shape} record)')
ax[1].set_ylabel('Numero di Studenti')

plt.tight_layout()
plt.show()
```
